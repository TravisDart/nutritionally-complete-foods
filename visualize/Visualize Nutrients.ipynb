```python
import matplotlib.pyplot as plt
import string

from ipywidgets import widgets, fixed, Layout

# fmt: off
foods = [
    ['14091', 'Beverages', 'Beverages, almond milk, unsweetened, shelf stable', '', (1840, 'mg'), (13, 'g'), (31, 'mg'), (0, 'mg'), (150, 'kcal'), (10, 'g'), (2, 'g'), (0, None), (10, 'µg'), (3, 'mg'), (60, 'mg'), (400, 'µg'), (1, 'mg'), (90, 'mg'), (670, 'mg'), (4, 'g'), (0, 'mg'), (1, 'µg'), (720, 'mg'), (0, 'mg'), (0, 'µg'), (0, 'mg'), (0, 'µg'), (0, 'mg'), (10, 'µg'), (63, 'mg'), (0, 'µg'), (965, 'g'), (1, 'mg')],
    ['14355', 'Beverages', 'Beverages, tea, black, brewed, prepared with tap water', '', (0, 'mg'), (3, 'g'), (4, 'mg'), (0, 'mg'), (10, 'kcal'), (0, 'g'), (0, 'g'), (3730, 'µg'), (50, 'µg'), (0, 'mg'), (30, 'mg'), (2190, 'µg'), (0, 'mg'), (10, 'mg'), (370, 'mg'), (0, 'g'), (0, 'mg'), (0, 'µg'), (30, 'mg'), (0, 'mg'), (0, 'µg'), (0, 'mg'), (0, 'µg'), (0, 'mg'), (0, 'µg'), (0, 'mg'), (0, 'µg'), (997, 'g'), (0, 'mg')],
    ['9068', 'Fruits and Fruit Juices', "Cherries, sour, red, frozen, unsweetened (Includes foods for USDA's Food Distribution Program)", '', (130, 'mg'), (110, 'g'), (56, 'mg'), (1, 'mg'), (460, 'kcal'), (4, 'g'), (16, 'g'), (0, None), (50, 'µg'), (5, 'mg'), (90, 'mg'), (570, 'µg'), (1, 'mg'), (160, 'mg'), (1240, 'mg'), (9, 'g'), (0, 'mg'), (0, 'µg'), (10, 'mg'), (0, 'mg'), (440, 'µg'), (1, 'mg'), (0, 'µg'), (17, 'mg'), (0, 'µg'), (1, 'mg'), (15, 'µg'), (872, 'g'), (1, 'mg')],
    ['11672', 'Vegetables and Vegetable Products', 'Potato pancakes', '', (320, 'mg'), (278, 'g'), (742, 'mg'), (2, 'mg'), (2677, 'kcal'), (148, 'g'), (33, 'g'), (4, 'µg'), (410, 'µg'), (17, 'mg'), (360, 'mg'), (2600, 'µg'), (17, 'mg'), (1280, 'mg'), (6220, 'mg'), (61, 'g'), (2, 'mg'), (89, 'µg'), (7640, 'mg'), (2, 'mg'), (320, 'µg'), (4, 'mg'), (3, 'µg'), (276, 'mg'), (3, 'µg'), (2, 'mg'), (27, 'µg'), (478, 'g'), (7, 'mg')],
]
# fmt: on

foods = [[y[0] for y in x[4:]] for x in foods]

dietary_requirements = [
    ["Calcium, Ca", (1300000, "mg"), (3000000, "mg"), "mg"],
    ["Carbohydrate", (465000, "g"), (671000, "g"), "g"],
    ["Choline", (550000, "mg"), (3000000, "mg"), "mg"],
    ["Copper, Cu", (890, "mg"), (8000, "mg"), "mg"],
    ["Energy", (4132000, "kcal"), (None, None), "kcal"],
    ["Total lipid (fat)", (115000, "g"), (161000, "g"), "g"],
    ["Fiber, total dietary", (38000, "g"), (None, None), "g"],
    ["Fluoride, F", (3000000, "µg"), (10000000, "µg"), "µg"],
    ["Folate, total", (400000, "µg"), (800000, "µg"), "µg"],
    ["Iron, Fe", (11000, "mg"), (45000, "mg"), "mg"],
    ["Magnesium, Mg", (410000, "mg"), (None, None), "mg"],
    ["Manganese, Mn", (2200000, "µg"), (9000000, "µg"), "µg"],
    ["Niacin", (16000, "mg"), (30000, "mg"), "mg"],
    ["Phosphorus, P", (1250000, "mg"), (4000000, "mg"), "mg"],
    ["Potassium, K", (3000000, "mg"), (None, None), "mg"],
    ["Protein", (77000, "g"), (None, None), "g"],
    ["Riboflavin", (1300, "mg"), (None, None), "mg"],
    ["Selenium, Se", (55000, "µg"), (400000, "µg"), "µg"],
    ["Sodium, Na", (1500000, "mg"), (None, None), "mg"],
    ["Thiamin", (1200, "mg"), (None, None), "mg"],
    ["Vitamin A, RAE", (900000, "µg"), (2800000, "µg"), "µg"],
    ["Vitamin B-6", (1300, "mg"), (80000, "mg"), "mg"],
    ["Vitamin B-12", (2400, "µg"), (None, None), "µg"],
    ["Vitamin C, total ascorbic acid", (75000, "mg"), (1800000, "mg"), "mg"],
    ["Vitamin D (D2 + D3)", (15000, "µg"), (100000, "µg"), "µg"],
    ["Vitamin E (alpha-tocopherol)", (15000, "mg"), (800000, "mg"), "mg"],
    ["Vitamin K (phylloquinone)", (75000, "µg"), (None, None), "µg"],
    ["Water", (3700000, "g"), (None, None), "g"],
    ["Zinc, Zn", (11000, "mg"), (34000, "mg"), "mg"],
]


def component_multiply(foods, quantities):
    """
    Input:
    foods = [ # 4 x 29 matrix
        [1180, 92, 21, 0, 470, 10, 3, 0, 20, 2, 110, 2820, 4, 560, 270, 3, 1, 22, 390, 0, 630, 0, 6, 0, 10, 5, 2, 893, 1],
        [0, 3, 4, 0, 10, 0, 0, 3730, 50, 0, 30, 2190, 0, 10, 370, 0, 0, 0, 30, 0, 0, 0, 0, 0, 0, 0, 0, 997, 0],
        [80, 120, 55, 1, 456, 1, 36, 0, 10, 2, 60, 2670, 1, 110, 800, 5, 0, 1, 20, 0, 30, 1, 0, 140, 0, 13, 50, 873, 1],
        [390, 170, 555, 0, 1310, 50, 12, 0, 290, 5, 150, 790, 10, 900, 1760, 44, 2, 61, 2820, 1, 570, 1, 3, 37, 5, 3, 5, 723, 5]
    ]
    quantities = [1, 2, 3, 4]  # 1 x 4 matrix

    Output: 29 x 4 matrix
    """
    result = []
    for i in range(len(foods[0])):
        subresult = [0] * 4
        # print(dietary_requirements[i][0])
        for j in range(len(quantities)):
            # foods[j][4:] = [0] * len(foods[j][4:])
            # print(foods[j][i], quantities[j], foods[j][i] * quantities[j])
            subresult[j] = foods[j][i] * quantities[j]
        result.append(subresult)
        # print()

    return result


def convert_to_percentages(component_data):
    """
    Note dietary_requirements is a global variable
    """
    for i in range(len(component_data)):
        max_value_for_nutrient = dietary_requirements[i][1][0]
        component_data[i] = [x / max_value_for_nutrient for x in component_data[i]]

    return component_data


def stacked_bars(use_percentages: bool = False, **kwargs):
    """
    Creates a stacked bar chart for each list in a list of lists.

    Args:
      data: A list of lists, where each sublist represents a set of values to be stacked.
      labels: A list of labels for the data sets.
    """
    labels = [x[0] for x in dietary_requirements]
    nutrient_quantities = list(kwargs.values())
    data = component_multiply(foods, nutrient_quantities)
    colors = ["b", "g", "r", "c", "m", "y"]

    fig, ax = plt.subplots()  # Create the figure and axis
    ax.invert_yaxis()
    ax.set_ylabel("Nutrients")
    ax.set_title(
        "Nutrients from Foods\n"
        "The thin black line indicates the acceptable nutrient range.\n"
        "Short grey bars inidicate nutrients with a lower bound only.",
        fontsize=10,
    )

    if use_percentages:
        ax.set_xlabel("Percentages")
        ax.set_xlim(0, 6)
        data = convert_to_percentages(data)
    else:
        # The logarithmic scale allows all of the bars to be visible,
        # but it makes it look like the first food provides the majority of the nutrients.
        ax.set_xscale("log")
        ax.set_xlabel("Values")

    fig.set_figheight(0.35 * len(data))

    for j in range(len(data)):
        # Draw the bars representing the nutrient values.
        baseline = 0
        for i in range(len(nutrient_quantities)):
            bar = ax.barh(labels[j], data[j][i], left=baseline, color=colors[i])
            baseline += data[j][i]

        # Draw the bar representing the acceptable nutrient range.
        if use_percentages:
            # The max as a percentage of the min
            if dietary_requirements[j][2][0] is None:
                color = "#999999"
                width = 0.1
            else:
                color = "k"
                minimum = 1
                maximum = dietary_requirements[j][2][0] / dietary_requirements[j][1][0]
                width = maximum - minimum
        else:
            # As absolute values:
            minimum = dietary_requirements[j][1][0]
            maximum = dietary_requirements[j][2][0]
            if dietary_requirements[j][2][0] is None:
                color = "#999999"
                width = minimum + 1
            else:
                color = "k"
                width = maximum - minimum

        ax.barh(labels[j], width, left=minimum, color=color, height=0.35)

    plt.tight_layout()
    plt.show()


slider_init_values = [
    ["Almond milk (14091)", 1227],
    ["Tea (14355)", 804],
    ["Cherries (9068)", 3382],
    ["Potato pancakes (11672)", 913],
]
sliders = [
    widgets.IntSlider(
        description=slider[0],
        value=slider[1],
        min=0,
        max=3500,
        style={"description_width": "200px"},
        layout=Layout(width="500px"),
    )
    for slider in slider_init_values
]


# interact won't take a list, so construct this awkward dict
slider_dict = {string.ascii_lowercase[i]: z for i, z in enumerate(sliders)}

use_percentages_checkbox = widgets.Checkbox(    
    description="Use percentages", value=True
)  # Create a checkbox

out = widgets.interact(
    stacked_bars, use_percentages=use_percentages_checkbox, **slider_dict
)
```
